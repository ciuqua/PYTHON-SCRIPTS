```python
import pandas as pd
import datetime as dt

# modify csv file putting column names in the first row
# note that integers cannot have coma separators (removed them by formating number differently)

#STEPS has been recorded in STEPS directory pay attention on the syntax in brackets 
steps=pd.read_csv(r'C:\Users\ciuqu\PYTHON\PYTHON-SCRIPTS\STEPS\steps_csv.csv')
steps
```

```python
steps.columns
steps.info()
```

```python
# change 'PARTICIPANT' column to string
# change 'DATE' column to datetime

steps['PARTICIPANT'].astype(str)
steps['DATE'] = pd.to_datetime(steps['DATE'])
steps.info()
```

```python
# df['weekday'] = df['Timestamp'].dt.dayofweek
steps['weekday']= steps['DATE'].dt.dayofweek
steps
```

```python
# Participants mean weekday steps = pmwds

pmwds=steps['STEPS'].groupby([steps['PARTICIPANT'], steps['weekday']]).mean()

# print the grouped data

print(pmwds)
```
